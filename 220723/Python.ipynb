```python
def cal(num1, num2, op):
  if op == '+':
    return num1 + num2
  elif op == '-':
    return num1 - num2
  elif op == '*':
    return num1 * num2
  elif op == '/':
    return num1 / num2
  else :
    return num1 % num2

num1= int(input("수를 입력 해 주세요."))
num2 = int(input("수를 입력해주세요")) 
op = input("연산자는?")

print(cal(num1, num2, op))
```

```python
class Person:
  def __init__(self, n, a):
    self.name = n
    if 0 <= a <= 20:
      self.__age = a
    else:
        self.__age = 0
  def inc_age(self):
    self.__age += 1
  def info(self):
    print(self.__age)

person = Person("Zangsu", 25)
person.inc_age()
person.info()
```

```python
#클래스 속성
class Person:
  count = 0 # 모든 인스턴스가 공유하는 클래스 속성
  def __init__(self, name):
    self.name = name
    Person.count += 1

person1 = Person('Zangsu')
print(person1.count)
person2 = Person('허준')
print(person1.count)
```

```python
#정적 메소드
class Math:
  @staticmethod
  def add(a, b):
    return a+b
  
  @staticmethod
  def sub(a, b):
    return a-b

print(Math.add(4, 5))
```

```python
class Rect:
  def __init__(self, h, v):
    self.h = h
    self.v = v
  def area(self):
    return self.h * self.v
  def inc(self):
    self.h *= 2
    self.v *= 2
    
rect1 = Rect(2, 3)
print(rect1.area())
rect1.inc()
print(rect1.area())
```

```python
class Vehicle:
  def __init__(self, speed):
    self.speed = speed
  def speed_up(self):
    self.speed+= 10
  def speed_dn(self):
    self.speed -= 10

class Car(Vehicle):
  def __init__(self, speed, wheels, seats):
    Vehicle.__init__(self, speed)
    self.wheels = wheels
    self.seats = seats
  def info(self):
    print(self.speed, self.wheels, self.seats)

class Truck(Car):
  def __init__(self, speed, loadge):
    Car.__init__(self, speed, 4, 4)
    self.loadge = loadge
  def load(self):
    print("load")
  def unload(self):
    print("unload")
  def info(self):
    Car.info(self)
    print(self.loadge)
  
truck = Truck(80, 50)
truck.load()
truck.info()
```

```python
#람다
# 람다함수 : 딱 한번 사용할 간단한 함순데 형식적인 타이핑은 번거로울 수 있다. 간단하게 일회용 함수를 정의하자.
# 표현 식 안에서 새로운 변수를 선언할 수 없다.

f = lambda x, y : x+y # '이름' = lambda '매개변수들' : '리턴값'

print(f(1,2))
```

```python
# map함수

nums = [1,2,3,4,5]
def double(x):
  return x**2

print(list(map(double, nums)))
print(list(map(lambda x : x**2, nums)))

nums2 = [6,7,8,9,10]

print(list(map(lambda x, y: x*y, nums, nums2)))
```

```python
#filter : 값이 true가 나오면 값을 반환, false면 반환하지 않음? ㅇㅇ

ages = [18, 19, 39, 12, 43, 13, 21, 25]

def adult_func(age):
  if age >= 19:
    return True
  else:
    return False

for age in filter(adult_func, ages):
  print(age)
print()
for age in filter(lambda age: age>=19, ages):
  print(age)
```
