## Series
* 1차원 배열과 같은구조
* 값을 저장하는 value와 색인를 저장하는 index로 구성
* 동일한 종류의 자료를 저장

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
s = pd.Series(np.random.randn(5))
print(s)
```

```
s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
print(s)
```

```
dic = {'a' : 0, 'b' : 1, 'c' : 2}
series1 = pd.Series(dic)
print(series1)
```

```
pd.Series(dic, index=['b', 'c', 'd', 'a'])
```

```
dates = pd.date_range('20220724', periods = 10)
print(dates)
```

* NaN (not a number) 결측치 - 값이 없거나 형식이 안맞는 경우


## DataFrame

* 엑셀과 같이 여러가지의 데이터가 모인 자료구조
* 여러개의 Series가 모인 자료구조

```
df = pd.DataFrame(np.random.randn(10,4), index=dates, columns=list('ABCD'))
print(df)
```

```
df2 = pd.DataFrame({'A': 1.,
                    'B': pd.Timestamp('20130102'),
                    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
                    'D': np.array([3]*4, dtype='int32'),
                    'E':['test', 'train', 'test', 'train'],
                    'F': 'foo'})
print(df2)
```
