# Matplotlib
* 파이썬용 시각화 라이브러리 이다.
* 라인플롯, 바 차트, 파이차트, 히스토그램 등의 차트와 플롯 등 다양한 스타일 지원
* 머신러닝시 데이터 분석이나 결과 확인을 위한 시각화에 사용

### matplotlib를 import 해서 간단한 Line Chart 그리기

* 코드를 간결하게 사용하기 위해서 matplotlib의 pyplot을 import하고 plt라고 alias를 설정

* plt.plot([x데이터], y데이터)

* '[ ]'(대괄호)안에 있는 내용은 생략가능, x데이터가 생략될 경우에 x데이터는 0부터 y데이터의 갯수만큼 1씩증가



```python
from matplotlib import pyplot as plt
plt.plot([1,2,3,4], [4,8,6,1]) #plot은 점을 잇는 직선을 긋는 것이기 때문에, x좌표와 y좌표가 필요하다.
plt.show()
```

```python
plt.plot([3, 8, 2, 5]) #x축 값이 생략이 되었을 떄는 0부터 1씩 증가하는 값이 기본으로 들어감
plt.show()
```

### 차트 제목, 축 이름 설정

```python
plt.plot(["Seoul", "Daegu", "Busan", "Jinju"], [30, 34, 12, 15])
plt.xlabel('City')
plt.ylabel('Income')
plt.title('Experiment Result')
plt.show()
```

### 복수의 데이터사용

```python
plt.plot([1,2,3],[1,4,2])
plt.plot([2,3,4], [3,7,8]  )
plt.xlabel('Sequence')
plt.ylabel('Time(sec)')
plt.title('Experiment Result')
plt.show()
```

### 범례(legend) 설정

```python
#위의 그래프에서 각 그래프가 무슨 그래픈지 구분이 안된다. 이를 위해 범례(legend)를 설정하자.
index = list(range(1,6))
powerNum = list(map(lambda x : x**2, index))
bin = list(map(lambda x : 2**x, index))
plt.plot(index, powerNum)
plt.plot(index, bin)
plt.legend(['x^2', '2^x'])
plt.show()
```

### plot함수의 매개변수들

* plot(\*args, scalex=True, scaley=True, data=None, \**kwargs)
* 매개변수에 \* 하나인것은 tuple
* 매개변수에 \** 두개인것은 dictionary

#### args

plot([x데이터1], y데이터1, [fmt], [x데이터2], y데이터2, [fmt], [x데이터3], y데이터3, [fmt],.....)

[]가 붙어있는 매개변수는 생략 가능하다.

```python

plt.plot([1,3,2,4], [3,4,5,6], ':gv', [1,2,3,4], [5,8,3,2], '-o', [1,2,3,4], [7,9,2,3], 'bv--', [1,2,3,4],[3,3,7,3], '*-.c')
plt.legend(['one', 'two', 'three', 'four'])
plt.show()
```

### subplot
* 두개이상의 차트가 필요할때 사용

```python
import numpy as np

x1= np.linspace(0.0, 5.0)
x2= np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2*np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('cos(x) * e^(-x)')

plt.subplot(2,1, 2)
plt.plot(x2, y2)
plt.title('cos(x)')

plt.show()
```

### Bar chart

```python
y = [5,3,7,9,3,5,6,7,8,9,2,3,4]
x= range(len(y))
print(x)
plt.bar(x,y, width=0.5, color = 'red')
```

### Pie chart

```python
labels = ['Samsung', 'Apple', 'LG', 'Huawei', 'Zangsu']
sizes = [34, 25, 13, 5, 100]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explodes = (0.2, 0.1, 0, 0, 0.1) #떨어진 정도
plt.title("Market Share Ratio")
plt.pie(sizes, labels = labels, colors = colors, autopct='%1.2f%%', startangle=90, explode = explodes, shadow = True)
plt.show()
```

### scatter (산포도)

```python
x=np.random.normal(0,1, 500)
y = np.random.normal(0, 1, 500)
plt.title('Scatter Plot')
plt.scatter(x, y, alpha = 0.5)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```
